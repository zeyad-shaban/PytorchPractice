```python
import torch
import imageio
import os
```

```python
img_paths = []
t = torch.tensor([1,2])
for filename in os.listdir("./data/images"):
    if(os.path.splitext(filename)[-1] == ".jpg"):
        img_paths.append(os.path.join("./data/images", filename))

img_paths
```

```python
batch_size = len(img_paths)
batch = torch.zeros(batch_size, 3, 256, 256)
```

```python
for i, img_path in enumerate(img_paths):
    batch[i] = torch.from_numpy(imageio.imread(img_path)).permute(2, 0, 1)[:3].contiguous()

mean = batch.mean(dim=[0,2,3], keepdim=True)
std = batch.std(dim=[0,2,3], keepdim=True)

batch = (batch - mean) / std
batch
```
