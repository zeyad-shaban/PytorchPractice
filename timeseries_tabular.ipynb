```python
dir = "./data/timeSeries/hour-fixed.csv"
```

```python
data_np = np.loadtxt(dir, dtype=np.float32, delimiter=",", skiprows=1, converters={1: lambda x: float(x[8:10])})
bikes = torch.from_numpy(data_np)
```

```python
daily_weather = bikes.view(-1, 24, bikes.shape[1])
daily_weather.transpose_(1,2)
```
