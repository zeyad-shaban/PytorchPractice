```python
import torch
import numpy as np
import re
```

```python
with open("./book.txt", encoding="utf-8") as f:
    text = f.read().lower()

ascii_values = torch.tensor(list(text.encode("utf-8")))
ascii_values
```

```python
text_onehot = torch.zeros(ascii_values.shape[0], 256, dtype=torch.bool)
text_onehot.scatter_(1, ascii_values.unsqueeze(-1).long(), True)
```

```python
text = re.sub(r'[^\w\s]', '', text)
text = re.sub(r'\s+', ' ', text).strip()
words = text.split(' ')
```

```python
words_list = sorted(set(words))
words2idx_dict = {word: i for (i,word) in enumerate(words_list)}

words2idx_dict
```

```python
text_onehot_word = torch.zeros(len(words), len(words2idx_dict), dtype=torch.bool)
for i, word in enumerate(words):
    text_onehot_word[i][words2idx_dict[word]] = True

text_onehot_word
```
